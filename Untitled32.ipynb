```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, add
import matplotlib.pyplot as plt
import urllib.request
import urllib.error
import os

# Load pre-trained ResNet50 model + higher level layers
base_model = ResNet50(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.layers[-2].output)

# Parameters
max_length = 34  # maximum length of the caption
vocab_size = 10000  # size of the vocabulary

# Load an image from URL
def load_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img

# Extract features from the image
def extract_features(img_path):
    img = load_image(img_path)
    features = model.predict(img, verbose=0)
    return features

# Simple tokenizer and padded sequences
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<unk>', filters='!"#$%&()*+.,-/:;=?@[\\]^_`{|}~ ')
tokenizer.fit_on_texts(['a simple example caption'])  # Example to initialize the tokenizer
word_index = tokenizer.word_index

# Simple caption generation model
def define_captioning_model(vocab_size, max_length):
    inputs1 = Input(shape=(2048,))
    fe1 = Dense(256, activation='relu')(inputs1)
    inputs2 = Input(shape=(max_length,))
    se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
    se2 = LSTM(256)(se1)
    decoder1 = add([fe1, se2])
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)
    model = Model(inputs=[inputs1, inputs2], outputs=outputs)
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    return model

captioning_model = define_captioning_model(vocab_size, max_length)

# Dummy training (this example does not actually train the model)
# You would normally train the model on a large dataset like MS COCO
captioning_model.summary()

# Example inference code to generate captions
def generate_caption(photo):
    in_text = 'startseq'
    for i in range(max_length):
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = captioning_model.predict([photo, sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = tokenizer.index_word.get(yhat, '<unk>')
        in_text += ' ' + word
        if word == 'endseq':
            break
    return in_text

# Function to check if the URL is reachable
def is_url_reachable(url):
    try:
        with urllib.request.urlopen(url) as response:
            return response.status == 200
    except:
        return False

# Function to prompt user for a valid URL
def get_valid_url():
    while True:
        url = input("Please provide a valid image URL: ")
        if is_url_reachable(url):
            return url
        else:
            print("The provided URL is not reachable. Please try again.")

# Valid URL for demonstration
img_url = 'https://tse4.mm.bing.net/th/id/OIP.cIKstCBiDCQtdLVGuskwxAHaE8?rs=1&pid=ImgDetMain'

# Check if URL is reachable and download image
if not is_url_reachable(img_url):
    img_url = get_valid_url()

try:
    urllib.request.urlretrieve(img_url, 'example.jpg')
    features = extract_features('example.jpg')
    caption = generate_caption(features)

    # Display image and generated caption
    plt.imshow(image.load_img('example.jpg'))
    plt.title(caption)
    plt.show()
except urllib.error.HTTPError as e:
    print(f"HTTP Error: {e.code} - {e.reason}")
except urllib.error.URLError as e:
    print(f"URL Error: {e.reason}")
except Exception as e:
    print(f"Unexpected error: {e}")
```
